### Segment 5 - Tokenization

```python
import nltk
nltk.download('punkt')
```

```python
text = 'On Wednesday, the Association for Computing Machinery, the world’s largest society of computing professionals, announced that Drs. Hinton, LeCun and Bengio had won this year’s Turing Award for their work on neural networks. The Turing Award, which was introduced in 1966, is often called the Nobel Prize of computing, and it includes a $1 million prize, which the three scientists will share.'
```

**Sentence tokenization**

```python
from nltk.tokenize import sent_tokenize
sent_tk = sent_tokenize(text)
print(f"Sentence tokenizing the text: \n")
print(sent_tk)
```

```python
from nltk.tokenize import word_tokenize
words = word_tokenize(text)
print(words)
```

**Removing _stop_ words**


```python
nltk.download('stopwords')
sw = set(stopwords.words("english"))
print(f"Stop words in English: {sw}")
```

```python
filtered_words = [w for w in words if not w in sw]
print(f"Text after removing stopwords:\n {filtered_words}")
```

**Stemming**


```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
```

```python
stemmed_words = []
for w in filtered_words:
    stemmed_words.append(stemmer.stem(w))
```

```python
print(f"Filtered words: \n{filtered_words}")
print(f"Stemmed words: \n{stemmed_words}")
```

**Lemmatizing words**

```python
lemm_words = []
for i in range(len(filtered_words)):
    lemm_words.append(lem.lemmatize(filtered_words[i]))
print(lemm_words)
```

**Parts of speech tagging**

```python
nltk.download("averaged_perceptron_tagger")
from nltk import pos_tag
```
