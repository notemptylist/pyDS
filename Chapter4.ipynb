### Segment 2 - Object Oriented Plot Construction

```python
%matplotlib inline
rcParams['figure.figsize'] = 6,5
```

**Defining axes, ticks and grids**

```python
x = range(1, 10)
y = [1, 2, 3, 4, 0, 4, 3, 2, 1]

fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])
ax.plot(x, y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])
ax.set_xlim([1, 9])
ax.set_ylim([0, 5])
ax.set_xticks(list(range(11)))
ax.set_yticks(list(range(6)))
ax.plot(x, y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])
ax.set_xlim([1, 9])
ax.set_ylim([0, 5])
ax.set_xticks(list(range(11)))
ax.set_yticks(list(range(6)))
ax.grid()
ax.plot(x, y)
```

**Generating multiple plots in one figure with subplots**

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1, 2) 
ax1.plot(x)
ax2.plot(x, y)
```

### Segment 3 - Plot formatting

**Defining plot colors**

```python
x = range(1, 10)
y = [1, 2, 3, 4, 0.5, 4, 3, 2, 1]

plt.bar(x, y)
```

```python
wide = [.5, .5, .5, .9, .3, .9, .5, .5, .5]
color = ['salmon']
plt.bar(x, y, width=wide, color=color, align='center')
```

```python
df = cars[['cyl', 'mpg', 'wt']]
df.plot()
```

```python
color_theme = ['darkgray', 'lightsalmon', 'powderblue']
df.plot(color=color_theme)
```

```python
z = [1, 2, 3, 4, 5]
color_theme = ['#665566', '#3388AA', '#A9A9A9', '#B0E0E6', '#cc33cc']
plt.pie(z, colors=color_theme)
plt.show()
```

**Customizing line styles**

```python
x1 = range(0, 10)
y1 = range(10, 0, -1)
plt.plot(x, y, ds='steps', lw=5)
plt.plot(x1, y1, ls='--', lw=10)
plt.show()
```

**Set plot markers**


```python
plt.plot(x, y, marker='+', mew=20)
```

### Segment 4 - Creating labels and annotations

**Labeling plot features**

The functional methdod

```python
x = range(1, 10)
y = [1,2,3,4,.5,4,3,2,1]
plt.bar(x, y)
plt.xlabel('your x-axis label')
plt.ylabel('your y-axis label')
plt.show()
```

```python
z = [1, 2, 3, 4, 5]
veh_type = ['bicycle', 'motorbike', 'car', 'van', 'stroller']
plt.pie(z, labels=veh_type)
plt.show()
```

```python
mpg = cars.mpg
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars.name, rotation=60, fontsize='medium')
ax.set_title('miles per gallon of cars in mtcars dataset')
ax.set_xlabel('car names')
ax.set_ylabel("MPG")
```

**Adding a legend to your plot**

```python
plt.pie(z)
plt.legend(veh_type, loc='best')
plt.show()
```
